```python
import pandas as pd
```

```python
airbnb_search_details = pd.read_csv('data/airbnb_search_details.csv')
airbnb_search_details = airbnb_search_details[['city', 'property_type', 'bathrooms', 'bedrooms']]
airbnb_search_details.head()
```

```python
result = airbnb_search_details.groupby(['city', 'property_type']).mean(['bathrooms', 'bedrooms']).reset_index()
result
```
