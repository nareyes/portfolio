Compare each employee's salary with the average salary of the corresponding department. Output the department, first name, and salary of employees along with the average salary of that department.

https://platform.stratascratch.com/coding/9917-average-salaries?code_type=1

```python
import pandas as pd
```

```python
employee = pd.read_csv('data/salesforce_employee.csv')
employee.head()
```

```python
employee['avg_salary'] = employee.groupby(['department'])['salary'].transform('mean')
employee.head()
```

```python
result = employee[['department', 'first_name', 'salary', 'avg_salary']]
result = result.sort_values(['department', 'first_name'])
result
```
